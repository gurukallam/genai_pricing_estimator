```python
# Data representing OpenAI GenAI model costs over time for different models (fictional and approximated for demonstration)
# Pricing data is based on approximate trends, not exact historical values
%pip install pandas matplotlib

import pandas as pd
import matplotlib.pyplot as plt

data_costs = {
    "Year": [2020, 2021, 2022, 2023, 2024],
    "GPT-3 (Cost per 1k tokens)": [0.06, 0.05, 0.045, 0.04, 0.035],  # Hypothetical decrease over years for GPT-3
    "GPT-3.5 (Cost per 1k tokens)": [None, None, 0.04, 0.035, 0.03],  # Started in 2022
    "GPT-4 (Cost per 1k tokens)": [None, None, None, 0.06, 0.05]      # Started in 2023
}

# Convert the data to a DataFrame
df_costs = pd.DataFrame(data_costs)

# Plotting the line chart for each model cost over time
plt.figure(figsize=(10, 6))

# Plot each series, taking care of possible None values
plt.plot(df_costs["Year"], df_costs["GPT-3 (Cost per 1k tokens)"], marker='o', linestyle='-', label="GPT-3")
plt.plot(df_costs["Year"], df_costs["GPT-3.5 (Cost per 1k tokens)"], marker='o', linestyle='--', label="GPT-3.5")
plt.plot(df_costs["Year"], df_costs["GPT-4 (Cost per 1k tokens)"], marker='o', linestyle=':', label="GPT-4")

# Chart customization
plt.title("OpenAI GenAI Model Costs Over Time")
plt.xlabel("Year")
plt.ylabel("Cost per 1k Tokens (USD)")
plt.legend(title="Model")
plt.grid(True)
plt.ylim(0, 0.07)  # Set y-axis limit to 0.07 USD for clearer view

plt.show()
```
