```python
# Re-import necessary libraries and re-define the data due to environment reset
import matplotlib.pyplot as plt
import pandas as pd

# Creating fictional data for GPU and TPU costs and adoption over time
# This data is illustrative and does not reflect exact historical pricing or adoption trends

data_gpu_tpu = {
    "Year": [2010, 2012, 2014, 2016, 2018, 2020, 2022, 2024],
    "GPU Cost per Unit (USD)": [800, 750, 700, 650, 600, 550, 500, 450],
    "TPU Cost per Unit (USD)": [None, None, None, 700, 600, 500, 450, 400],  # TPUs start being available around 2016
    "GPU Adoption (Units in millions)": [10, 20, 35, 50, 70, 90, 110, 130],
    "TPU Adoption (Units in thousands)": [None, None, None, 5, 20, 40, 60, 85]
}

# Convert data to a DataFrame
df_gpu_tpu = pd.DataFrame(data_gpu_tpu)

# Plotting the line chart with two y-axes for costs and adoption trends
fig, ax1 = plt.subplots(figsize=(12, 8))

# Plot GPU and TPU costs on the left y-axis
ax1.set_xlabel("Year")
ax1.set_ylabel("Cost per Unit (USD)", color="tab:blue")
ax1.plot(df_gpu_tpu["Year"], df_gpu_tpu["GPU Cost per Unit (USD)"], marker='o', color="tab:blue", label="GPU Cost")
ax1.plot(df_gpu_tpu["Year"], df_gpu_tpu["TPU Cost per Unit (USD)"], marker='o', linestyle="--", color="tab:cyan", label="TPU Cost")
ax1.tick_params(axis="y", labelcolor="tab:blue")
ax1.legend(loc="upper left", title="Cost Trend")

# Create a second y-axis for GPU and TPU adoption rates
ax2 = ax1.twinx()  
ax2.set_ylabel("Adoption (Units)", color="tab:green")
ax2.plot(df_gpu_tpu["Year"], df_gpu_tpu["GPU Adoption (Units in millions)"], marker='o', color="tab:green", label="GPU Adoption")
ax2.plot(df_gpu_tpu["Year"], df_gpu_tpu["TPU Adoption (Units in thousands)"], marker='o', linestyle="--", color="tab:olive", label="TPU Adoption")
ax2.tick_params(axis="y", labelcolor="tab:green")
ax2.legend(loc="upper right", title="Adoption Trend")

plt.title("GPU and TPU Costs and Adoption Trends Over Time (2010-2024)")
fig.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
