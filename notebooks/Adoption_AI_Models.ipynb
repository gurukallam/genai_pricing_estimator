```python
# Re-import necessary libraries and re-define the data due to environment reset
import matplotlib.pyplot as plt
import pandas as pd

# Fictional data representing the adoption of different types of models over time (illustrative purposes)
data_model_adoption = {
    "Year": [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    "Classification Models": [20, 25, 30, 40, 50, 60, 65, 70, 72, 75],
    "Regression Models": [10, 15, 20, 25, 30, 40, 45, 50, 52, 55],
    "Generative Models": [5, 10, 20, 30, 45, 60, 75, 85, 90, 95]
}

# Convert data to DataFrame
df_model_adoption = pd.DataFrame(data_model_adoption)

# Plotting the line chart for model adoption trends over time
plt.figure(figsize=(10, 6))

# Plot each series
plt.plot(df_model_adoption["Year"], df_model_adoption["Classification Models"], marker='o', linestyle='-', label="Classification Models")
plt.plot(df_model_adoption["Year"], df_model_adoption["Regression Models"], marker='o', linestyle='--', label="Regression Models")
plt.plot(df_model_adoption["Year"], df_model_adoption["Generative Models"], marker='o', linestyle='-.', label="Generative Models")

# Chart customization
plt.title("Adoption Trends of Different AI Model Types Over Time")
plt.xlabel("Year")
plt.ylabel("Adoption Rate (%)")
plt.legend(title="Model Type")
plt.grid(True)
plt.ylim(0, 100)  # Set y-axis limit for clearer comparison

plt.show()
```
